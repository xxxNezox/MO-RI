Свёрточные нейросети используют операцию свёртки для обработки входных данных. В случае изображений свёртка представляет собой процесс применения фильтра (или ядра) к изображению для извлечения признаков.

1 Сохранение пространственной структуры
2 Параметры фильтров (ядер)
3 Параметрическая эффективность

Он не обрабатывает все пиксели изображения сразу, а только часть изображения, что позволяет ему фокусироваться на локальных признаках.

Свёрточные нейросети могут автоматизировать этот процесс, обучая фильтры, которые извлекают похожие признаки (градиенты, текстуры) без необходимости ручной настройки.
В свёрточных нейросетях аналогичная задача решается через несколько слоёв свёртки и пулинга

```python
import numpy as np
from skimage.feature import local_binary_pattern
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('./data/stereo.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

radius = 1
n_points = 8

lbp_image = local_binary_pattern(image, n_points, radius, method='uniform')

plt.imshow(lbp_image, cmap='gray')
plt.title('LBP Image')
plt.show()
```
