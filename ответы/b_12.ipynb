2. Используйте OpenCV для реализации алгоритма, который строит трехмерную 
модель объекта на основе стереопары изображений. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

stereo_img = cv2.imread("./data/stereo2.jpeg", cv2.IMREAD_GRAYSCALE)

height, width = stereo_img.shape
left_img = stereo_img[:, :width // 2]
right_img = stereo_img[:, width // 2:]

stereo = cv2.StereoBM_create()
disparity = stereo.compute(left_img, right_img)

plt.imshow(disparity, cmap='plasma')
plt.colorbar()
plt.show()
```
