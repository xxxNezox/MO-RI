ССегментация изображений —это процесс разделения изображения на множество 
сегментов (набор пикселей, или связанные области), чтобы упростить или изменить 
представление изображения в более значимую и легкую для анализа форму.


Методы сегментации изображений
1. Пороговая сегментация - разделение изображения на сегменты на 
основе интенсивности пикселей (Простой и быстрый метод, идеален для 
изображений с высоким контрастом между объектом и фоном).
2. Кластеризация - группировка пикселей на основе их свойств, таких 
как цвет, интенсивность или текстура (Эффективна для более сложных 
изображений, где пороговая сегментация не дает четких результатов).
3. Сегментация на основе контуров - выделение границ объектов на 
изображении для определения их формы и положения (Подходит для 
задач, где важно точно определить форму объекта).

SSR — это метод компенсации неравномерного освещения, который улучшает видимость деталей в тенях и переосвещённых областях.
(Сегментация медицинских изображений, где неравномерное освещение может скрывать важные детали.)

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def cross_entropy_loss(y_true, y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)) # на этом моменте я умер

def forward_pass(X, weights):
    return sigmoid(np.dot(X, weights))

def backward_pass(X, y_true, y_pred):
    error = y_pred - y_true
    gradient = np.dot(X.T, error) / len(X)
    return gradient

def update_weights(weights, gradient, learning_rate):
    return weights - learning_rate * gradient
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(
    n_samples=1000, 
    n_features=20,
    n_classes=2,
    random_state=1
)
y = y.reshape(-1, 1)

np.random.seed(42)
input_size = X.shape[1]
output_size = 1 
weights = np.random.randn(input_size, output_size) * 0.01 

learning_rate = 0.01
epochs = 1000

for epoch in range(epochs):
    y_pred = forward_pass(X, weights)
    
    loss = cross_entropy_loss(y, y_pred)
    
    gradient = backward_pass(X, y, y_pred)
    
    weights = update_weights(weights, gradient, learning_rate)
    
    if epoch % 100 == 0:
        print(f"Эпоха {epoch}, Потери: {loss:.4f}")

y_pred = forward_pass(X, weights)
print("Предсказания модели:")
print(y_pred)
```
