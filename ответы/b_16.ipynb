```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('./data/RANSAC/RANSAC1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

filtered_image = cv2.bilateralFilter(image, d=25, sigmaColor=150, sigmaSpace=150)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Исходное изображение')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(filtered_image)
plt.title('Сглаженное изображение (билатеральный фильтр)')
plt.axis('off')

plt.show()
```
